```python
# Importing the required libraries
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# load the data
iris = pd.read_csv('./data/Iris.csv')
iris.head()
```

```python
iris.shape
```

```python
iris = iris.drop('Id',axis=1)
iris.head()
```

```python
iris.describe()
```

```python
# count samples per class
iris['Species'].value_counts()
```

```python
(iris['SepalLengthCm'].mean(),iris['SepalWidthCm'].mean(),iris['PetalLengthCm'].mean(),iris['PetalWidthCm'].mean())
```

```python
(iris['SepalLengthCm'].median(),iris['SepalWidthCm'].median(),iris['PetalLengthCm'].median(),iris['PetalWidthCm'].median())
```

```python
iris.var()
```

```python
iris.std()
```

```python
# quartiles /  quantiles 10%
iris.quantile(0.1)
```

```python
iris.quantile(0.50)
```

```python
# inter quartile range IQR

iris.quantile(0.75) - iris.quantile(0.25)
```

```python
iris.groupby(['Species'])['PetalLengthCm'].mean()
```

```python
# visualizing with matplotlib and seaborn

iris.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',figsize=(8,6),color='red',marker='o',s=50)
```

```python
sb.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=iris)
```

```python
sb.FacetGrid(iris,hue='Species').map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

```python
sb.FacetGrid(iris,hue='Species').map(plt.scatter,'PetalLengthCm','PetalWidthCm').add_legend()
```

```python
# Violin plot using Matplot libs and Sb:  for distribution of data

plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sb.violinplot(x='Species',y='SepalLengthCm',data=iris)
plt.subplot(2,2,2)
sb.violinplot(x='Species',y='SepalWidthCm',data=iris)
plt.subplot(2,2,3)
sb.violinplot(x='Species',y='PetalLengthCm',data=iris)
plt.subplot(2,2,4)
sb.violinplot(x='Species',y='PetalWidthCm',data=iris)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sb.boxplot(x='Species',y='SepalLengthCm',data=iris)
plt.subplot(2,2,2)
sb.boxplot(x='Species',y='SepalWidthCm',data=iris)
plt.subplot(2,2,3)
sb.boxplot(x='Species',y='PetalLengthCm',data=iris)
plt.subplot(2,2,4)
sb.boxplot(x='Species',y='PetalWidthCm',data=iris)
```

```python
sb.boxplot(x='Species',y='SepalWidthCm',data=iris)
sb.stripplot(x='Species',y='SepalWidthCm',data=iris, color='r', jitter=True)
```
